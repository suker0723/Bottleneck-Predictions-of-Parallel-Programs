```python
import numpy as np
import matplotlib.pyplot as plt
```

## plot matrix size speedup for simple matrix multiplication

```python
m100 = [1, 0.8347940403, 0.8004201681]
m200 = [1, 0.9831747881, 0.7876509928]
m500 = [1, 0.9690277512, 1.003777716]
m800 = [1, 1.023715169, 0.9953554526]
m1000 = [1, 0.9961292407, 1.00427649]
thread_label = [1, 2, 4]
x = np.arange(len(thread_label))  # the label locations
```

```python
cmap=['b','g','r','c','m','y','k','blueviolet','chartreuse']
def fit(ax,xdata,ydata,cmap,cindex,label):
    ax.scatter(xdata,ydata,color=cmap[cindex])
    ax.plot(xdata,ydata,color=cmap[cindex],label=label)
    for i in range(len(xdata)):
        ax.annotate('%.2f'%ydata[i], (xdata[i], ydata[i]))

fig,ax=plt.subplots(1,1,figsize=(20,8)) # figure size
plt.rcParams.update({'font.size': 30}) # font size
label=""
data_frame=None

fit(ax, x, m100,cmap,0, "100")
fit(ax, x, m200,cmap,1, "200")
fit(ax, x, m500,cmap,2,  "500")
fit(ax, x, m800,cmap,3, "800")
fit(ax, x, m1000,cmap,4, "1000")
ax.set_xticks(x, thread_label)
ax.legend()

plt.xticks(fontsize=18, rotation=0)
plt.yticks(fontsize=18, rotation=0)
ax.set_ylabel("Speedup",fontsize=30) # y labels
ax.set_xlabel("Num of Threads",fontsize=30) # x labels
plt.suptitle("SpeedUp vs Thread")
```

## Relationship between cache miss rate and matrix size

```python
m100 = [0.50, 0.50, 0.50]
m200 = [0.48, 0.48, 1.77]
m500 = [4.62, 4.62, 4.62]
m800 = [4.63, 4.63, 4.63]
m1000 = [4.63, 4.63, 4.63]
thread_label = [1, 2, 4]
x = np.arange(len(thread_label))
```

```python
cmap=['b','g','r','c','m','y','k','blueviolet','chartreuse']
def fit(ax, xdata, ydata, cmap, cindex, label):
    bar_width = 0.10
    offset = bar_width * (cindex - 2)
    ax.bar([x + offset for x in xdata], ydata, width=bar_width, color=cmap[cindex], label=label)
    for i in range(len(xdata)):
        ax.annotate('%.2f' % ydata[i], (xdata[i] + offset, ydata[i]))

fig,ax=plt.subplots(1,1,figsize=(20,12)) # figure size
plt.rcParams.update({'font.size': 18}) # font size
label=""
data_frame=None

fit(ax, x, m100,cmap,0, "100")
fit(ax, x, m200,cmap,1, "200")
fit(ax, x, m500,cmap,2,  "500")
fit(ax, x, m800,cmap,3, "800")
fit(ax, x, m1000,cmap,4, "1000")
ax.set_xticks(x, thread_label)
ax.legend()

plt.xticks(fontsize=18, rotation=0)
plt.yticks(fontsize=18, rotation=0)
ax.set_ylabel("Percentage",fontsize=30) # y labels
ax.set_xlabel("Num of Threads",fontsize=30) # x labels
plt.suptitle("Overall Cache Miss Rate in level 1 Data Read")
```
